```python
%pip install ultralytics
import ultralytics
ultralytics.checks()
import cv2
from ultralytics import YOLO
import numpy as np
import pandas as pd
import glob
import os
```

Next we will load the model. You may find other models [here](https://github.com/ultralytics/ultralytics/tree/main/ultralytics/models/v8)

We included this line as a seperate function - this way you can look at the documentation and change it to try other models.

Press the play button below to continue

```python
model = YOLO('yolov8n-pose.pt')
```

<p>Finally - run the model and any videos you have in the folder!</p>
<p>Congratulations - you have completed your first exercise using a model built from machine learning.</p> 
You can check the 'Your_ML_Analysis' folder on the left for your video and image results.

```python
output_directory = '/content/Your_ML_Analysis'
output_filename = 'PoseOverlay.mp4'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Construct the output file path
output_path = os.path.join(output_directory, output_filename)
count = 0
video_lib = glob.glob('/content/*.mp4') + glob.glob('/content/*.MOV')
#video_path = "/content/Curl_Rep2.mp4"
for file_path in video_lib:
  video_path = str(file_path)
  cap = cv2.VideoCapture(video_path)
  fourcc = cv2.VideoWriter_fourcc(*'mp4v')
  video = cv2.VideoCapture(video_path)
  ret, frame = video.read()
  height, width, _ = frame.shape
  baseline = "/content/Your_ML_Analysis/Baseline.jpg"
  
  out = cv2.VideoWriter("/content/Your_ML_Analysis/PoseOverlay.mp4", fourcc, 30.0, (frame.shape[1], frame.shape[0]))
  while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()
    if success:
      # Run YOLOv8 inference on the frame
      results = model.predict(source = frame, conf=.6)
      annotated_frame = results[0].plot()
      out.write(annotated_frame)
      if count < 1:
          cv2.imwrite(baseline, annotated_frame)
      else:
        break
    else:
        # Break the loop if the end of the video is reached
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
